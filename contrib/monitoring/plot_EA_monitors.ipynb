# Plotting EA Monitor
In this notebook we plot the data collected in the printout of the EA monitor. Such printout is a text file named RUN\<timestapt\>.eaMonitor that DenoptimGA creates and updates ruding an evolutionary experiment. Note that such monitoring was introdiced from DENOPTIM version 3.0.

Change the following pathname to be the pathname of the monitor file you want to plot:


```python
monitorFile ='/tmp/denoptim_Diamond/RUN20211123103149.eaMonitor'
```

```python
import pandas as pd

data = pd.read_csv(monitorFile, delim_whitespace=True)
```

```python
data.columns
```

```python
import matplotlib.pyplot as plt

plt.plot(data['XOVERATTEMPTS'])
plt.plot(data['MUTATTEMPTS'])
plt.plot(data['BUILDANEWATTEMPTS'])

plt.legend(['XOVERATTEMPTS','MUTATTEMPTS','BUILDANEWATTEMPTS'])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot(data['XOVERATTEMPTS']-data['FAILEDXOVERATTEMPTS'])
plt.plot(data['XOVERATTEMPTS'])
plt.plot(data['FAILEDXOVERATTEMPTS'])
plt.plot(data['DUPLICATEPREFITNESS'])

plt.legend(['XOVERATTEMPTS-FAILEDXOVERATTEMPTS','XOVERATTEMPTS','FAILEDXOVERATTEMPTS','DUPLICATEPREFITNESS'])
plt.show()
```
